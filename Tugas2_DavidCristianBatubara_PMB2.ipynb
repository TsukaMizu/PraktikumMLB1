```python
%matplotlib inline
from copy import deepcopy
import numpy as np
import pandas as pd
import seaborn as sns
from mpl_toolkits.mplot3d import Axes3D
from sklearn.cluster import KMeans
from sklearn.datasets import make_blobs
from matplotlib import pyplot as plt
from sklearn.impute import SimpleImputer
plt.rcParams['figure.figsize'] = (16, 9)
plt.style.use('ggplot')
```

```python
penguins = sns.load_dataset('penguins')
penguins
```

```python
f1 = penguins['bill_length_mm'].values
f2 = penguins['bill_depth_mm'].values
X = np.array(list(zip(f1, f2)))
plt.scatter(f1, f2, c='black', s=7)
```

```python
# Menghapus baris dengan NaN di X
X = X[~np.isnan(X).any(axis=1)]

# Lakukan K-Means clustering
kmeans = KMeans(n_clusters=3)
kmeans = kmeans.fit(X)
labels = kmeans.predict(X)
C = kmeans.cluster_centers_

# Cetak nilai centroid
print(C)
```

```python
plt.scatter(X[:, 0], X[:, 1], s=7, c=labels)
plt.scatter(C[:, 0], C[:, 1], marker='*', s=200, c='#050505')
```

```python
# Men-generate dataset yang terkelompok dalam 4 cluster
X, y = make_blobs(n_samples=800, n_features=3, centers=4)
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.scatter(X[:, 0], X[:, 1], X[:, 2])
```

```python
# Initializing KMeans
kmeans = KMeans(n_clusters=4)
# Fitting with inputs
kmeans = kmeans.fit(X)
# Predicting the clusters
labels = kmeans.predict(X)
# Getting the cluster centers
C = kmeans.cluster_centers_
```

```python
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.scatter(X[:, 0], X[:, 1], X[:, 2], c=y)
ax.scatter(C[:, 0], C[:, 1], C[:, 2], marker='*', c='#050505', s=1000)
```

```python
for k in range (1, 10):
  # Menentukan jumlah cluster
  kmeans = KMeans(n_clusters=k, random_state=1)
  # Fitting input data
  kmeans = kmeans.fit(X)
  # Mendapatkan cluster labels
  labels = kmeans.predict(X)
  # Menghitung jumlahan jarak antara setiap sampel dengan cluster centroid-nya (SSE)
  inertia = kmeans.inertia_
  print(f"k: {k}, cost: {inertia}")
```

```python
from sklearn.metrics.cluster import silhouette_score
silhouette_score(X, labels)
```

```python
# Pilih hanya fitur numerik dari dataset penguins
features = penguins[['bill_length_mm', 'bill_depth_mm']].dropna().values
# Data X sudah berupa array 2D yang hanya berisi nilai numerik
X = features  # Tidak perlu zip atau reshape lagi

# Inisialisasi list untuk menyimpan nilai SSE dan silhouette scores
SSE = []
silhouette_scores = []

# Tentukan rentang jumlah cluster K
K = range(1, 10)

# Lakukan K-Means clustering untuk berbagai nilai K
for k in K:
    kmeanModel = KMeans(n_clusters=k, random_state=42)
    kmeanModel.fit(X)  # Hanya sekali memanggil fit

    # Hitung Sum of Squared Errors (SSE) atau inertia_
    SSE.append(kmeanModel.inertia_)

    # Hitung silhouette score jika k > 1
    if k > 1:
        score = silhouette_score(X, kmeanModel.labels_)
        silhouette_scores.append(score)
```

```python
# Plot the elbow
plt.plot(K, SSE, 'bx-')
plt.xlabel('k')
plt.ylabel('SSE')
plt.title('The Elbow Method showing the optimal k')
plt.show()
```

```python
# Plot the Silhouette Coefficient
plt.figure(figsize=(8, 5))
plt.plot(range(2, 10), silhouette_scores, 'go-')
plt.xlabel('Jumlah Cluster (K)')
plt.ylabel('Silhouette Coefficient')
plt.title('Silhouette Coefficient untuk Setiap K')
plt.show()
```

```python
# Membuat figure dan axis untuk subplot
figure, ax = plt.subplots(2, 3, figsize=(15, 10))  # 2 baris, 3 kolom

K = range(1, 7)  # Jumlah cluster dari 1 hingga 6

for k in K:
    # Menentukan posisi subplot
    if k < 4:
        row = 0
        column = k - 1
    else:
        row = 1
        column = k - 4

    # Melakukan clustering KMeans
    kmeanModel = KMeans(n_clusters=k, random_state=42)
    y_pred = kmeanModel.fit_predict(X)

    # Membuat plot scatter
    ax[row][column].scatter(X[:, 0], X[:, 1], c=y_pred, cmap='winter_r')
    ax[row][column].set_title(f'K = {k}')
    ax[row][column].set_xlabel('Bill Length (mm)')
    ax[row][column].set_ylabel('Bill Depth (mm)')

plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from sklearn.datasets import make_blobs

# Membuat dataset contoh dengan 100 sampel dan 2 fitur
X, _ = make_blobs(n_samples=100, centers=4, random_state=42)

# Inisialisasi dengan 'k-means++' (default)
kmeans_plus = KMeans(n_clusters=4, init='k-means++', random_state=42)
kmeans_plus.fit(X)
labels_plus = kmeans_plus.predict(X)

# Inisialisasi dengan 'random'
kmeans_random = KMeans(n_clusters=4, init='random', random_state=42)
kmeans_random.fit(X)
labels_random = kmeans_random.predict(X)

plt.figure(figsize=(12, 6))
# Visualisasi hasil dengan 'k-means++'
plt.subplot(1, 2, 1)
plt.scatter(X[:, 0], X[:, 1], c=labels_plus, cmap='viridis', s=50)
plt.scatter(kmeans_plus.cluster_centers_[:, 0], kmeans_plus.cluster_centers_[:, 1],
            s=200, c='red', marker='X')
plt.title("Inisialisasi: k-means++")

# Visualisasi hasil dengan 'random'
plt.subplot(1, 2, 2)
plt.scatter(X[:, 0], X[:, 1], c=labels_random, cmap='viridis', s=50)
plt.scatter(kmeans_random.cluster_centers_[:, 0], kmeans_random.cluster_centers_[:, 1],
            s=200, c='red', marker='X')
plt.title("Inisialisasi: Random")

plt.show()
```
