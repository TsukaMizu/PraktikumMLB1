```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs
from scipy.cluster.hierarchy import dendrogram, linkage
from scipy.spatial.distance import pdist, squareform
from scipy.cluster.hierarchy import linkage, fcluster
```

### Generate Random data

```python
x1, y1 = make_blobs(n_samples=50, centers=[[2,5],[3,6],[4,2]], cluster_std=0.9, random_state=42)
```

```python
plt.scatter(x1[:,0], x1[:,1], marker='o')
```

```python
distance_matrix = pdist(x1)
distance_matrix = squareform(distance_matrix)

print(distance_matrix)
```

```python
# Single Linkage
single_linkage = linkage(x1, method='single')

# Average Linkage
average_linkage = linkage(x1, method='average')
```

```python
def plot_dendrogram(linkage_matrix, labels, **kwargs):
    dendrogram(linkage_matrix, labels=labels, **kwargs)

#Single Linkage
plt.figure(figsize=(12, 6))
plt.subplot(1, 2, 1)
plot_dendrogram(single_linkage, labels=[str(i) for i in range(len(x1))])
plt.title('Single Linkage')

# Average Linkage
plt.subplot(1, 2, 2)
plot_dendrogram(average_linkage, labels=[str(i) for i in range(len(x1))])
plt.title('Average Linkage')
```

### Penjelasan

- Single Linkage menggabungkan dua cluster berdasarkan jarak terdekat antara titik-titik di kedua cluster, cenderung menghasilkan cluster yang memanjang dan lebih sensitif terhadap outlier

- Average Linkage menggabungkan dua cluster berdasarkan rata-rata jarak antara semua pasangan titik di kedua cluster, menghasilkan cluster yang lebih seimbang dan stabil, serta kurang dipengaruhi oleh outlier

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
from sklearn.datasets import load_diabetes
```

```python
diabetes = load_diabetes()
X = diabetes.data
y = diabetes.target
```

```python
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

```python
pca = PCA(n_components=2)
X_pca = pca.fit_transform(X_scaled)
```

```python
plt.figure(figsize=(8, 6))
plt.scatter(X_pca[:, 0], X_pca[:, 1], c=y, cmap='viridis', edgecolor='k', s=50)
plt.title('PCA pada informasi tentang pasien diabetes', fontsize=16)
plt.colorbar(label='Target (Progression of Diabetes)')
plt.show()
```

```python
print(f"Variansi yang dijelaskan oleh masing-masing komponen: {pca.explained_variance_ratio_}")
print(f"Total variansi yang dijelaskan oleh 2 komponen pertama: {sum(pca.explained_variance_ratio_)}")
```
